## 2. Przygotowanie danych tekstowych

```python
spam_dataset = pd.read_csv('spam.csv', encoding = "ISO-8859-1", usecols=[0, 1], names=['Spam', 'Text'],
                           skiprows=1)
spam_dataset['Spam'] = spam_dataset['Spam'].map({'ham': 0, 'spam': 1}).astype(int)
spam_dataset
```

```python
print(spam_dataset['Spam'].value_counts(normalize=True))
```

### Usuwanie znaków interpunkcyjnych
'!"#$%&\'()*+,-./:;&lt;=>?@[\\]^_`{|}~'

```python
def remove_puncation(text):
    cleaned = ''.join([word for word in text if word not in string.punctuation])
    return cleaned
spam_dataset['Cleaned_Text'] = spam_dataset['Text'].apply(lambda x: remove_puncation(x))
spam_dataset
```

### Tokenizacja

```python
import shutil
import os
nltk_dir = os.path.expanduser('~/nltk_data')
if os.path.exists(nltk_dir):
    shutil.rmtree(nltk_dir)

import nltk

# Usuń cache punkt tokenizera
import functools
if hasattr(nltk.tokenize.punkt, '_get_punkt_tokenizer'):
    nltk.tokenize.punkt._get_punkt_tokenizer.cache_clear()

# Pobierz ponownie wszystkie potrzebne zasoby
nltk.download('punkt', force=True)
nltk.download('stopwords', force=True)
nltk.download('punkt_tab', force=True)  # Specjalnie dla brakującego zasobu

print(nltk.data.path)

nltk.data.path.append('C:/Users/User/nltk_data')  # lub inna ścieżka z komunikatu błędu

from nltk.tokenize import RegexpTokenizer

def tokenize(text):
    tokenizer = RegexpTokenizer(r'\w+')
    return tokenizer.tokenize(text.lower())

import string
from nltk.corpus import stopwords

def clean_text(text):
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.lower()
    stop_words = set(stopwords.words('english'))
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

# Czyszczenie tekstu
spam_dataset['Cleaned_Text'] = spam_dataset['Text'].apply(clean_text)

# Tokenizacja przy użyciu nowej funkcji tokenize
spam_dataset['Tokenized_Text'] = spam_dataset['Cleaned_Text'].apply(lambda x: tokenize(x))
spam_dataset.head()
```

```python
def remove_stopwords(text):
    without_stopwords = [word for word in text if word not in stopwords]
    return without_stopwords
spam_dataset['WithoutStop_Text'] = spam_dataset['Tokenized_Text'].apply(lambda x: remove_stopwords(x))
spam_dataset
```

### Stemming

```python
stemmer = nltk.PorterStemmer()
def stemming(text):
    stemmed_words = [stemmer.stem(word) for word in text]
    return stemmed_words
spam_dataset['Stemmed_Text'] = spam_dataset['WithoutStop_Text'].apply(lambda x: stemming(x))
spam_dataset
```

### Lematyzacja

```python
lemmater = nltk.WordNetLemmatizer()
def lemmatizing(text):
    lemmatized_words = [lemmater.lemmatize(word) for word in text]
    return lemmatized_words
spam_dataset['Lemmatized_Text'] = spam_dataset['WithoutStop_Text'].apply(lambda x: lemmatizing(x))
spam_dataset
```

```python
words_spam = list(spam_dataset.loc[spam_dataset['Spam']==1, 'Lemmatized_Text'].values)
words_spam = list(itertools.chain.from_iterable(words_spam))
words_spam = ' '.join(words_spam)
words_notspam = list(spam_dataset.loc[spam_dataset['Spam']==0, 'Lemmatized_Text'].values)
words_notspam = list(itertools.chain.from_iterable(words_notspam))
words_notspam = ' '.join(words_notspam)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Wektoryzacja tekstu
count_v2 = CountVectorizer(max_df=0.1, max_features=500)
X_count_v2 = count_v2.fit_transform(spam_dataset['Lemmatized_Text'].apply(lambda x: ' '.join(x)))

# LDA
lda = LatentDirichletAllocation(n_components=7, random_state=2022,
                              learning_method='batch')
X_topics = lda.fit_transform(X_count_v2)

# Pobieranie nazw cech - NOWA SYNTAXA
feature_names = count_v2.get_feature_names_out()  # Zmienione na get_feature_names_out()

# Wyświetlanie tematów
for topic_idx, topic in enumerate(lda.components_):
    print(f'Temat: {topic_idx+1}')
    print(' '.join([feature_names[i] for i in topic.argsort()[:-5-1:-1]]))
    print()
```

```python
def vector_to_text(imdb_vector, label):
    reverse_index = {v:k for k,v in imdb.get_word_index().items()}
    return f"""Comment: {" ".join([reverse_index.get(i-3, "#") for i in imdb_vector])}
            Label: {"Positive" if label == 1 else "Negative"}"""
```

```python
# Poprawiona wersja - każdy print w osobnej linii bez niepotrzebnych wcięć
print(vector_to_text(x_train[0], y_train[0]))
print(vector_to_text(x_train[1], y_train[1])) 
print(vector_to_text(x_train[2], y_train[2]))
```

```python
# Robimy padding komentarzy tak, aby wszystekie miały tę sama długość
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
```

### 2. Trenowanie modelu

```python
## Tworzymy model sieci neuronowej z jedną warstwę ukryta z 16 węzłami (taki mamy rozmiar embeddingu)
def build_keras_model(input_dim, output_dim):
    model = Sequential()
    # Używamy tutaj regularyzacji L2, aby model nam nie overfitował
    model.add(Embedding(input_dim = input_dim, output_dim = output_dim, embeddings_regularizer=l2(0.01)))
    model.add(Flatten())
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
model = build_keras_model(num_words, embedding_dim)
```

```python
def encode_text(text):
    import re # import biblioteki do wyrazen regularnych
    def get_word_index(word): # Funkcja do zakodowania slowo za pomoca liczby z uwzglednieniem limitu slow
        w_idx = index.get(word, -1)
        return w_idx + 3 if w_idx <= num_words else 2
    index = imdb.get_word_index()
    text = re.sub(r'[^a-z ]', '', text.lower()) # Usuniecie wszystkich znakow, ktore nie sa literami oraz zamiana duzych liter na male
    encoded = [get_word_index(word) for word in text.split(" ")] # Zakodowanie wyrazow
    return pad_sequences([[1] + list(encoded)], maxlen=maxlen) # Dodanie znaku startu i ustawienie dlugosci na 200 słów
```
